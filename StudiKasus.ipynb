## Preprocessing, Scaling

### Tanpa library selain pandas,numpy, dan juga seaborn

Mengimpor data mentah yang masih ada data kosong kemudian diimputasi
Meng-encode string seperti di kolom beli dan negara
Kolom beli: 0 untuk Tidak, 1 untuk Iya
Kolom Negara: 0 Untuk Prancis, 1 Untuk Spanyol, 2 Untuk Germany


Kemudian membulatkan data dari usia dan gaji
Lalu membuat scaling Normalisasi
Dilanjutkan dengan membuat scaling Standarisasi

```python
#Kode 1
import pandas as pd
import numpy as np
import seaborn as sns

dataset = pd.read_csv(r"E:\datasets\datanegara\Data.csv")
print("Dataset Asli\n",dataset)
# Imputasi menggunakan mean untuk kolom Usia dan Gaji
dataset['Usia'].fillna(dataset['Usia'].mean(),inplace=True)
dataset['Gaji'].fillna(dataset['Gaji'].mean(),inplace=True)

# Encode kolom Beli: 'No' menjadi 0 dan 'Yes' menjadi 1
dataset['Beli'] = dataset['Beli'].map({'No': 0, 'Yes': 1})

dataset.loc[:,'Usia'] = dataset['Usia'].round()
dataset.loc[:,'Gaji'] = dataset['Gaji'].round()

# Encode kolom Negara ke angka
dataset['Negara'] = dataset['Negara'].map({'France':0,'Spain':1,'Germany':2})

# Tampilkan dataset setelah imputasi dan encoding
print("Dataset Imputasi\n",dataset)

print("")

#Normalisasi
# Normalisasi kolom Usia dan Gaji menggunakan Min-Max Normalization
dataset['Usia'] = (dataset['Usia'] - dataset['Usia'].min()) / (dataset['Usia'].max() - dataset['Usia'].min())
dataset['Gaji'] = (dataset['Gaji'] - dataset['Gaji'].min()) / (dataset['Gaji'].max() - dataset['Gaji'].min())

# Tampilkan dataset setelah normalisasi
print("Dataset Normalisasi\n",dataset)


#Standarisasi
mean_usia = np.mean(dataset['Usia'])
std_usia = np.std(dataset['Usia'])

mean_gaji = np.mean(dataset['Gaji'])
std_gaji = np.std(dataset['Gaji'])

dataset['Usia'] = (dataset['Usia']-mean_usia) / std_usia
dataset['Gaji'] = (dataset['Gaji']-mean_gaji) /std_gaji
print("")
print("Dataset Standarisasi\n",dataset)

#Regresi Linear Sederhana
```

```python
import numpy as np
import pandas as pd
import seaborn as sns

# Membaca dataset
dataset = pd.read_csv(r'E:\datasets\datagaji\Daftar_gaji.csv')

# Misalkan kolom 'Tahun_bekerja' adalah fitur (X) dan 'Gaji' adalah target (Y)
X = dataset['Tahun_bekerja'].values
Y = dataset['Gaji'].values

# Mengacak data dan membagi menjadi training set (80%) dan test set (20%)
np.random.seed(42)
indices = np.random.permutation(len(X))
train_size = int(len(X) * 0.8)
train_indices = indices[:train_size]
test_indices = indices[train_size:]

# Membagi data menjadi training dan testing
X_train, X_test = X[train_indices], X[test_indices]
Y_train, Y_test = Y[train_indices], Y[test_indices]

# Menghitung rata-rata X_train dan Y_train untuk regresi linear
mean_X = np.mean(X_train)
mean_Y = np.mean(Y_train)

# Menghitung slope (m) dan intercept (b) menggunakan rumus regresi linear
numerator = np.sum((X_train - mean_X) * (Y_train - mean_Y))
denominator = np.sum((X_train - mean_X) ** 2)
m = numerator / denominator  # Slope
b = mean_Y - m * mean_X  # Intercept

# Menampilkan hasil slope dan intercept
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")

# Membuat prediksi untuk data test
Y_pred = m * X_test + b

# Menambahkan hasil prediksi ke dalam DataFrame untuk visualisasi
test_results = pd.DataFrame({'Tahun_bekerja': X_test, 'Gaji Asli': Y_test.round(), 'Gaji Prediksi': Y_pred.round()})
print(test_results.head())

# Visualisasi hasil menggunakan hanya seaborn
sns.set_style("whitegrid")

# Mengurutkan data untuk visualisasi
sorted_indices = np.argsort(X_test)  # Mengurutkan indeks berdasarkan X_test
X_test_sorted = X_test[sorted_indices]
Y_pred_sorted = Y_pred[sorted_indices]
Y_test_sorted = Y_test[sorted_indices]

# Membuat scatter plot untuk data asli dan garis regresi
sns.scatterplot(x=X_test_sorted, y=Y_test_sorted, color='blue', label='Data Asli (Test Set)')  # Data asli test set
sns.lineplot(x=X_test_sorted, y=Y_pred_sorted, color='red', label='Garis Regresi')  # Garis regresi

# Menambahkan judul dan label
```

```python
import pandas as pd
import numpy as np
import seaborn as sns

# Membaca dataset dari file CSV
dataset = pd.read_csv(r"E:\pythonjupyter\coba.csv")

# Random Seed
np.random.seed(42)

# Menshuffle Dataset
shuffled_indices = np.random.permutation(dataset.index)
shuffled_df = dataset.loc[shuffled_indices]

# Menentukan ratio pembagian dataset, 0.8 berarti sebesar 80%
train_size = 0.8 
train_index = int(len(shuffled_df) * train_size)

# Membagi dataset menjadi 2 bagian 
train_set = shuffled_df.iloc[:train_index]
test_set = shuffled_df.iloc[train_index:]

# Menghitung rata-rata variabel x dan y
x_gaul = train_set["Tahun_bekerja"].mean()
y_gaul = train_set["Gaji"].mean()

# Menghitung Slope
atas = ((train_set['Tahun_bekerja'] - x_gaul) * (train_set['Gaji'] - y_gaul)).sum()
bawah = ((train_set['Tahun_bekerja'] - x_gaul) ** 2).sum()

slope = atas / bawah
intercept = y_gaul - slope * x_gaul

print("Nilai Slope {}", slope)
print("Nilai Intercept {}",intercept)

dataset['Predicted_Gaji'] = intercept + slope * dataset['Tahun_bekerja']

dataset["Predicted_Gaji"] = dataset["Predicted_Gaji"].round()

# Mengisi nilai kosong di kolom Gaji dengan Predicted Gaji
dataset['Gaji'].fillna(dataset['Predicted_Gaji'], inplace=True)

# Menghapus kolom Predicted_Gaji setelah diisi
dataset.drop(columns=['Predicted_Gaji'], inplace=True)

print(dataset)

korelasi = dataset['Tahun_bekerja'].corr(dataset['Gaji'])
print("Besaran korelasi:\n",korelasi)

sns.regplot(data=dataset, x='Tahun_bekerja', y='Gaji', color='Magenta', scatter_kws={'s': 50}, ci=0)
```
