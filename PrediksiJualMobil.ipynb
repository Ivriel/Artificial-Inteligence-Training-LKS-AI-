```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Preparation

```python
dataset = pd.read_csv(r'E:\pythonjupyter\datasets\car_data.csv')
dataset
```

# EDA

```python
print("Data Kosong:\n", dataset.isnull().sum())
print("\nData NaN:\n", dataset.isna().sum())
print("\nData Duplikat:\n", dataset.duplicated().sum())
print("\nTipe Data:\n", dataset.dtypes)
```

```python
rows, columns = dataset.shape
print("\nVariabel (Kolom):", columns)
print("Baris:", rows)
```

```python
dataset['Gender'] = dataset['Gender'].map({'Male': 0, 'Female': 1})

print("\nUnique Values in Gender Column:", dataset['Gender'].unique())
print("\nTipe Data Setelah Encoding:\n", dataset.dtypes)
print("\nData NaN Setelah Encoding:\n", dataset.isna().sum())
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Age', data=dataset)
plt.title("Boxplot of Age by Gender")
plt.xlabel("Gender (0 = Male, 1 = Female)")
plt.ylabel("Age")
plt.show()
```

```python
correlations = dataset.corr()
sns.heatmap(data=correlations, annot=True, fmt=".2f", cmap="coolwarm")
```
